## *1) Fonction d'insertion* ##

```python
import pymongo, csv
```

```python
from pymongo import MongoClient     #Ne pas oublier de Run mongod via le terminal#
```

```python
client = MongoClient() #First step créer un MongoClient pour le lier à l'instance mongod# 
# client = MongoClient ('localhost', numéroDuPort) du coup au dessus on est par défault#
```

```python
db = client.Floupics
collection = db.Movies
#Collections and databases are created when the first document is inserted into them.
```

```python
def insert_documents(csvFilePath):
    
    try:
        
        data = [] #si csv mettre liste meme si dictreader#
        
        with open(csvFilePath, encoding='utf-8') as csvf: 
            
            csvReader = csv.DictReader(csvf)
            
            for e in csvReader:
                data.append(e)
                
            
            collection.insert_many(data) #insert_one pour un seul document#
            #.inserted_id à voir mais une clé à été générée automatiquement sans la méthode#
        
        print("ah bah etonnement t'as réussis, bah gg mon gars")
        

    
    except Exception as a : 
        
        print(a)
        
        print("comme attendu, t'as fais d'la merde")
        
        
    return


        
```

## *2) Insertion Movies* ##

```python
chemin = r'CSV/movies.csv'

insert_documents(chemin)
```
